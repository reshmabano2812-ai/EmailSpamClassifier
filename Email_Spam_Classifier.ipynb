IMPORTING LIBRARIES...

```python
import numpy as py
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
```

LOADING DATA...

```python
df=pd.read_csv('mail_data.csv')
```

DATA PREPROCESSING...

```python
data=df.where((pd.notnull(df)), ' ')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['Category'] == 'spam' ,'Category']=0
data.loc[data['Category'] == 'ham', 'Category']=1
```

```python
X= data['Message']
Y= data['Category']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING OF TRAINING AND TESTING DATASET

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

VECTORIZIATION OF DATA...

```python
feature_extraction=TfidfVectorizer(min_df=1, stop_words='english' , lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

IMPLEMENTING MODEL

```python
model=MultinomialNB()
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training_data=model.predict(X_train_features)
accuracy=accuracy_score(X_train, prediction_on_training_data)
```

```python
print("="*50)
print(f"OVERALL MODEL ACCURACY:{accuracy_score(Y_train, prediction_on_training_data)*100:.2f}%")
print("\nCLASSIFICATION REPORT")
print(classification_report(Y_train, prediction_on_training_data))
print("="*50)
```

```python
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test, prediction_on_test_data)
```

```python
print("="*50)
print(f"OVERALL MODEL ACCURACY:{accuracy_score(Y_test, prediction_on_test_data)*100:.2f}%")
print("\nCLASSIFICATION REPORT")
print(classification_report(Y_test, prediction_on_test_data))
print("="*50)
```

```python
print('accuracy on test data :', accuracy_on_test_data)
```
